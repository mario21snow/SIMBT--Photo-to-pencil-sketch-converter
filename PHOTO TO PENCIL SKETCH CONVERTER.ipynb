## TASK 02 : PHOTO TO PENCIL SKETCH CONVERTER


* Image loading: Load the input image into the program.

* Grayscale conversion: Convert the input image into grayscale format to remove any color information.

* Invert colors: Invert the grayscale image to obtain the negative image.

* Gaussian blur: Apply a Gaussian blur to smooth the image and remove any noise.

* Edge detection: Use an edge detection algorithm (such as the Canny edge detector) to detect the edges in the image.

* Thresholding: Apply a thresholding operation to the edge-detected image to remove any weak edges and obtain a binary image.

* Dilation: Dilate the binary image to thicken the edges.

* Final sketch: Combine the dilated binary image with the negative grayscale image to obtain the final pencil sketch.

# import the library

```python
import cv2
import matplotlib.pyplot as plt
```

# get the image location and the image file name

```python
image = cv2.imread("li.jpg")
```

# Converting the image to RGB format
-----------------------------------------------------------------
## Matplotlib interprets images in RGB format, but OpenCV uses BGR format so to convert the image so that it's properly loaded, convert it before loading.

```python
li = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(li)
plt.axis(False)
plt.show()
```

# Converting the image to greyscale image

```python
grey_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale_image)
```

# Inverting the image

```python
inverted_image = cv2.bitwise_not(grey_scale_image)
plt.imshow(inverted_image)
```

# Blurring the image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (111,111), 0)
plt.imshow(blurred_image)
```

# Inverting the blurred image

```python
inverted_blur_image = cv2.bitwise_not(blurred_image)
plt.imshow(inverted_blur_image)
```

# Creating the pencil sketch using the grey scale and inverted image

```python
sketch_image = cv2.divide(grey_scale_image, inverted_blur_image, scale=256.0)
plt.imshow(sketch_image)
```

# Saving the image

```python
cv2.imwrite("sketch.png", sketch_image)
```

# Converting the image back to BRG format

```python
plt.figure(figsize = (10,10))
image3 = cv2.cvtColor(sketch_image, cv2.COLOR_RGB2BGR)
plt.imshow(image3)
plt.axis(False)
plt.show()
```
